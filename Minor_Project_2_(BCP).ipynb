```python
#BREAST CANCER PREDICTION
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
```

```python
#Load_Dataset
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df.head()
```

```python
#EDA
print('shape of dataset:',df.shape)
print(df['target'].value_counts()) #class distribution
print(df.describe())
```

```python
#missing value
print(df.isnull().sum())
print("-----------------------------")
print(df.isnull().sum().sum())
```

```python
#train test split
x = df.drop('target',axis=1)
y = df['target']
x_train,x_test ,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
#Feature Scaling
scalar = StandardScaler()
x_train_scaled = scalar.fit_transform(x_train)
x_test_scaled = scalar.transform(x_test)
```

```python
#Model Training
model = LogisticRegression()
model.fit(x_train_scaled,y_train)
```

```python
#Model Evaluation
y_pred = model.predict(x_test_scaled)
print("accuracy score:",accuracy_score(y_test , y_pred))
print("ClassificationReport:\n",classification_report(y_test,y_pred))
```

```python
#Confusion matrix
cm = confusion_matrix(y_test , y_pred)
sns.heatmap(cm,annot=True ,fmt = 'd',cmap='Blues' , xticklabels=['Malignant' , 'Benign'], yticklabels=['Malignant' , 'Benign'])
plt.xlabel('predicted')
plt.ylabel('Actual')
plt.title("Confusion Matrix")
```

```python
sample = x_test.iloc[0].values.reshape(1, -1)
sample_scaled = scalar.transform(sample)
prediction = model.predict(sample_scaled)
print("Prediction for sample 0:", "Benign" if prediction[0] == 1 else "Malignant")

```
